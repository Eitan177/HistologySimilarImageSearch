```
#@markdown ### Run The Code Cell To Instantiate the Random Forest


from os import walk
import re

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, f1_score
from sklearn.metrics import classification_report
from sklearn.decomposition import TruncatedSVD
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc

from scipy import interp
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import auc
from sklearn.model_selection import StratifiedKFold

def load_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, (299, 299))
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return img, image_path

image_model = tf.keras.applications.DenseNet121(include_top=False,
                                                weights='imagenet')
new_input = image_model.input
hidden_layer = image_model.layers[-1].output

image_features_extract_model = tf.keras.Model(new_input, hidden_layer) 


def to_bf_np(img_name_vector):
  encode_train = sorted(set([i[0] for i in img_name_vector if len(i)>0]))

  # Feel free to change batch_size according to your system configuration
  image_dataset = tf.data.Dataset.from_tensor_slices(encode_train)
  image_dataset = image_dataset.map(
    load_image, num_parallel_calls=tf.data.experimental.AUTOTUNE).batch(2)

  for img, path in image_dataset:
    batch_features = image_features_extract_model(img)
    batch_features = tf.reshape(batch_features,
                                (batch_features.shape[0], -1, batch_features.shape[3]))

    for bf, p in zip(batch_features, path):
      path_of_feature = p.numpy().decode("utf-8")
      np.save(path_of_feature, bf.numpy())
  return encode_train

# Load the numpy files
def map_func(img_name):
  img_tensor = np.load(img_name.decode('utf-8')+'.npy')
  return img_tensor



def ld_imgs(encode_train):
  dataset = tf.data.Dataset.from_tensor_slices(encode_train)

  dataset = dataset.map(lambda item1: tf.numpy_function(
            map_func, [item1], [tf.float32]),
            num_parallel_calls=tf.data.experimental.AUTOTUNE)
  return(dataset)


def name_of_files_through_graph(f):
  encode_train=to_bf_np(f)
  dataset=ld_imgs(encode_train)
  max_pool =to_mp(dataset)
  max_pool=tf.convert_to_tensor(np.stack(max_pool).squeeze())
  return([max_pool, f])

def to_mp(dataset):
  max_pool=[]
  for (batch, (img_tensor)) in enumerate(dataset):
    #max_pool = np.max(img_tensor, axis=1)
    img_tensor=tf.expand_dims(img_tensor,axis=0)
    max_pool.append(tf.nn.max_pool(img_tensor,img_tensor.shape[2],img_tensor.shape[2],'SAME'))
  return (max_pool)

def makemp(fname):
  f = []
  for (dirpath, dirnames, filenames) in walk(fname):
      print(dirpath)
      yy=[[os.path.join(dirpath,fns)] for fns in filenames if len(re.findall('npy$',fns))==0]
      print(yy)
      f.extend([[os.path.join(dirpath,fns)] for fns in filenames if len(re.findall('npy$',fns))==0])
      break
  max_p, f=name_of_files_through_graph(f)
  return max_p, f


def eval_model(features_s,labels_s):
  cv = StratifiedKFold(n_splits=6)

  X = features_s
  y = labels_s
  print(X)
  n_samples, n_features = X.shape
  classifier=forest_clf
  classifierType='Random Forest'
  tprs = []
  aucs = []
  mean_fpr = np.linspace(0, 1, 100)
  fig, ax = plt.subplots(1, 1, figsize = (15, 10))
  for i, (train, test) in enumerate(cv.split(X, y)):
    classifier.fit(X.numpy()[train], y[train])
    viz = plot_roc_curve(classifier, X.numpy()[test], y[test],
                                  name='ROC fold {}'.format(i),
                                  alpha=0.3, lw=1, ax=ax)
    interp_tpr = interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)


  ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
                  label='Chance', alpha=.8)
  mean_tpr = np.mean(tprs, axis=0)
  mean_tpr[-1] = 1.0
  mean_auc = auc(mean_fpr, mean_tpr)
  std_auc = np.std(aucs)
  ax.plot(mean_fpr, mean_tpr, color='b',
                  label=r'Mean ROC (AUC = %0.2f $\pm$ %0.2f)' % (mean_auc, std_auc),
                  lw=2, alpha=.8)

  std_tpr = np.std(tprs, axis=0)
  tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
  tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
  ax.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2,
                          label=r'$\pm$ 1 std. dev.')

  ax.set(xlim=[-0.05, 1.05], ylim=[-0.05, 1.05],
                title="Receiver operating characteristic example {}".format(classifierType))
  ax.legend(loc="lower right")
  return classifier  

n_estimators=100
max_depth= 5
max_leaf_nodes= 25
class_weights='balanced'
forest_clf = RandomForestClassifier(n_estimators=int(n_estimators), max_depth=int(max_depth),
                                        max_leaf_nodes=int(max_leaf_nodes),
                                        class_weight=class_weights, oob_score=True,
                                        random_state=0) 


img_maxpool_1, img_1_n=makemp('folder1_keepers')
img_maxpool_2, img_2_n=makemp('folder2_keepers')

max_p_1and2=tf.concat([img_maxpool_1,img_maxpool_2],axis=0)
file_img_ns = img_1_n+ img_2_n

features12=max_p_1and2
labels12=np.repeat([0,1],[img_maxpool_1.shape[0],img_maxpool_2.shape[0]])

modelfitted=eval_model(features12,labels12)
```
